# import & get article ID

```python
import requests, json, os
from bs4 import BeautifulSoup as BS

with open ('nhk.json', 'r') as f:
    ids = sorted(json.load(f).keys())
    print('oldest ID: %s' % ids[0])
    print('newest ID: %s' % ids[-1])
```

# Scrape

```python
total_dic = {}

for i in range(192320, 192325):
    url = f"https://www3.nhk.or.jp/nhkworld/th/news/{i}/"
    response = requests.get(url)
    response.encoding='utf-8'
    if response.status_code == 200:
        dic = {}
        soup = BS(response.text, "html.parser")
        data = soup.find('script',type="application/ld+json")
        data = json.loads(data.text)
        dic['headline'] = data['headline']
        dic['article'] = data['articleBody']
        dic['date'] = data['datePublished']
        dic['url'] = url
        total_dic[i] = dic

with open ('nhk_new.json', 'w') as f:
    with open ('nhk.json', 'r') as g:
        dic = json.load(g)
        for k,v in total_dic.items():
            if str(k) not in dic:
                dic[str(k)] = v
    total_list = sorted(dic.items(), key=lambda x:int(x[0]))
    new_dic = {k:v for k, v in total_list}
    json.dump(dic,f, indent=4, ensure_ascii=False)
```

## Delete & Rename

```python
ans = input('delete? [Y/N]: ')
if ans = 'y' or 'Y':
    os.remove('nhk.json')
    os.rename('nhk_new.json', 'nhk.json')
```
